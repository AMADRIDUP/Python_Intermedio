#Mapas de calor y mapas coropléticos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-10,10,1000)
y1=np.sin(x)
y2=np.cos(x)

plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(["Seno","Coseno"],loc=2)
plt.title("Mis gráficas de Matplotlib")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.show()
```

```python
def circulo(radio,angulo=360):
  import math
  import pandas as pd
  import matplotlib.pyplot as plt
  x=[]
  y=[]
  for i in range(angulo):
    x.append(radio*math.cos(i))
    y.append(radio*math.sin(i))
  df=pd.DataFrame({"x":x,"y":y})
  return plt.scatter(df.x,df.y)
```

```python
for i in range(10,100,10):
  circulo(i)
plt.show()
```

```python
import seaborn as sns
import geopandas as gpd
```

```python
cdmx=gpd.read_file("alcaldias.geojson")
cdmx.head()
```

```python
cdmx.columns
```

```python
cdmx=cdmx[['nomgeo','geometry']]
cdmx.head()
```

```python
fig, ax =plt.subplots(figsize=(10,10))
cdmx.plot(ax=ax,color="green")
plt.show()
```

```python
data=pd.read_csv("covid.csv")
data.head()
```

```python
data["ENTIDAD UNIDAD MEDICA"].value_counts()
```

```python
covid=data[data["ENTIDAD RESIDENCIA"]=="CIUDAD DE MÉXICO"]
covid.head()
```

```python
covid_alcadia=covid["MUNICIPIO RESIDENCIA"]
covid_alcadia.head()
```

```python
conteo=covid_alcadia.value_counts()
conteo.sort_index(inplace=True)
conteo=conteo.reset_index()
conteo
```

```python
cdmx.sort_values("nomgeo")
```

```python
verificar=np.where(cdmx["nomgeo"]==conteo["index"],cdmx["nomgeo"],conteo["index"])
len(verificar)
```

```python
# join the geodataframe with the cleaned up csv dataframe
merged = cdmx.set_index("nomgeo").join(conteo.set_index("index"))
merged.head()
```

```python
merged.columns=["geometry","Casos"]
merged
```

```python
merged.index.name="Alcaldia"
merged.sort_index()
```

```python
fig, ax = plt.subplots(1, figsize=(15, 8))
merged.plot(column="Casos", cmap="Blues", linewidth=0.8, ax=ax, edgecolor="0.8", legend=True)
ax.axis("off")
titulo=str("Casos activos COVID-19 en CDMX")
ax.set_title(titulo, fontdict={"fontsize": "20", "fontweight" : "3"})
fig.savefig(titulo+".png", dpi=300)
plt.show()
```

```python
merged.reset_index(inplace=True)
fig, ax = plt.subplots(1, figsize=(15, 8))
merged.plot(column="Casos", cmap="Blues", linewidth=0.8, ax=ax, edgecolor="0.8", legend=True)
ax.axis("off")
titulo=str("Casos activos COVID-19 en CDMX")
ax.set_title(titulo, fontdict={"fontsize": "20", "fontweight" : "3"})
merged.apply(lambda x: ax.annotate(s=x.Alcaldia, xy=x.geometry.centroid.coords[0], ha='center'),axis=1);
fig.savefig(titulo+".png", dpi=300)
plt.show()
```
